```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler

from google.colab import drive
drive.mount('/content/drive')
path = '/content/Titanic-Dataset.csv'
df = pd.read_csv(path)
```

```python
df.head(15)
```

```python
# ============================================
# TITANIC SURVIVAL PREDICTION USING TENSORFLOW
# ============================================

# ============================================
# IMPORT LIBRARIES
# ============================================

import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

print("\nFIRST 5 ROWS")
print(df.head())

# ============================================
# SELECT REQUIRED FEATURES
# ============================================

# Input Features
X = df[['Pclass', 'Age', 'Fare']]

# Target
y = df['Survived']

# Fill missing Age values with mean
X['Age'] = X['Age'].fillna(X['Age'].mean())

# ============================================
# NORMALIZATION
# ============================================

# Min-Max Scaling
scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)

print("\nNORMALIZED FEATURES")
print(X_scaled[:5])

# ============================================
# TRAIN TEST SPLIT
# ============================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================
# CREATE NEURAL NETWORK
# ============================================

model = tf.keras.Sequential([

    # Hidden Layer
    tf.keras.layers.Dense(
        units=2,
        activation='sigmoid',
        input_shape=(3,)
    ),

    # Output Layer
    tf.keras.layers.Dense(
        units=1,
        activation='sigmoid'
    )

])


model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ============================================
# MODEL SUMMARY
# ============================================

print("\nMODEL SUMMARY")
model.summary()

# ============================================
# TRAIN MODEL
# ============================================

history = model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=16,
    validation_split=0.1
)


loss, accuracy = model.evaluate(X_test, y_test)

print("\nTEST LOSS:", loss)
print("TEST ACCURACY:", accuracy)

# ============================================
# MAKE PREDICTIONS
# ============================================

predictions = model.predict(X_test)

print("\nPREDICTIONS")
print(predictions[:10])

# ============================================
# CONVERT PROBABILITIES TO 0/1
# ============================================

predicted_classes = (predictions > 0.5).astype(int)

print("\nPREDICTED CLASSES")
print(predicted_classes[:10])

# ============================================
# TEST WITH CUSTOM INPUT
# ============================================

# Example:
# Pclass = 1
# Age = 24
# Fare = 120

custom_input = np.array([[1, 24, 120]])

# Normalize using same scaler
custom_input_scaled = scaler.transform(custom_input)

prediction = model.predict(custom_input_scaled)

print("\nCUSTOM PASSENGER PREDICTION")
print(prediction)

if prediction[0][0] > 0.5:
    print("SURVIVED")
else:
    print("NOT SURVIVED")

# ============================================
# DISPLAY FINAL WEIGHTS
# ============================================

print("\nFINAL WEIGHTS")

for layer in model.layers:

    weights, biases = layer.get_weights()

    print("\nWEIGHTS:")
    print(weights)

    print("\nBIASES:")
    print(biases)

# ============================================
# END
# ============================================
```

```python
# ============================================
# SAVE MODEL
# ============================================

model.save("titanic_ann_model.keras")

print("Model Saved Successfully")
```
