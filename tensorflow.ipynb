```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
import matplotlib.pyplot as plt
from keras.layers import Dense
from tensorflow.keras.models import Sequential
x=np.array([
    [120,25],
    [85,30],
    [150,35],
    [95,22
     ]

])
x=x/200
y=np.array([1,0,1,0])
print("input shape",x.shape)
print("output shape",y.shape)
```

```python
#load iris data\\
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
target_names=iris.target_names
print(target_names)
```

```python
#Standard scalar
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.transform(x_test)
#define the model architecture
model= Sequential()
model.add(Dense(10,activation='relu',input_shape=(x_train.shape[1],)))
model.add(Dense(5,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history= model.fit(x_train,y_train,epochs=50,validation_split=0.1,verbose=1)
```

```python
test_loss,test_acc = model.evaluate(x_test,y_test)
print('Test Accuracy:',test_acc)
```

```python
import pandas as pd
from google.colab import drive
from sklearn.preprocessing import LabelEncoder, StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Mount Google Drive (if not already mounted)
drive.mount('/content/drive')

# Load the dataset
path = '/content/CarPrice_dataset.csv'
df = pd.read_csv(path)

print("Original DataFrame head:")
display(df.head())
print("\nChecking for null values:")
display(df.isnull().sum())

# Drop duplicates and clean 'CarName'
df.drop_duplicates(inplace=True)
df['CarName'] = df['CarName'].str.lower()
df['CarName'] = df['CarName'].str.replace(" ", "_")

print("\nDataFrame head after initial cleaning:")
display(df.head())
```

### Data Preprocessing: Label Encoding and Scaling

```python
# Identify categorical and numerical columns
categorical_cols = ['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem', 'CarName']
numerical_cols = ['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']

# Apply Label Encoding to categorical columns
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Apply StandardScaler to numerical columns
sc = StandardScaler()
for col in numerical_cols:
    df[col] = sc.fit_transform(df[[col]])

print("DataFrame head after Label Encoding and Standard Scaling:")
display(df.head())
```

### Prepare Data for Model Training

```python
# Define features (x) and target (y)
x = df.drop(['car_ID', 'price'], axis=1)
y = df['price']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f"x_train shape: {x_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

### Define and Train the Neural Network Model

```python
model = Sequential([
    # Input Layer + Hidden Layer 1
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    # Hidden Layer 2
    Dense(32, activation='relu'),
    # Output Layer
    Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

model.summary()

history = model.fit(x_train, y_train, epochs=50, verbose=1, batch_size=32, validation_split=0.2)
```

### Evaluate Model Performance

```python
# Make predictions on the test data
predictions_on_test_data = model.predict(x_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, predictions_on_test_data)
mae = mean_absolute_error(y_test, predictions_on_test_data)
r2 = r2_score(y_test, predictions_on_test_data)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'R-squared (R2) Score: {r2}')
```

### Visualize Training History

```python
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()
```
